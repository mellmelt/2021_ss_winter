### pop()  
- 리스트중 가장 오른쪽에 있는 객체(마지막 항목)를 반환한후 삭제
```
test1 = [1,2,3]
test1.pop()
test1
```



- pop(i)  
    - 리스트중 인덱스 i 위치의 객체를 반환한후 삭제
    - 객체 x가 없으면 indexError가 발생


````python
test1 = [1,2,3,10,50,100]
p = test1.pop()
test1
p # 삭제한 항목 반환
p1 = test1.pop()
print(test1, p, p1)

p2 = test1.pop(1) # 제거할 위치 선정 가능 # 해당 위치 항목이 없으면 Error 발생
print(test1, p, p1, p2)
````

## split()
- 문자열 나누기
- 기준(빈칸,:, , . )등 기준으로 문자열을 나누어준다,
- 나누어진 문자열은 ***리스트***가 된다.

```
s = '문제 해결을 위한  파이썬' 
s.split()
```
- 결과
    - ['문제', '해결을', '위한', '파이썬']

````python
s = '인천대학교 국어국문학과 홍길동' 
#학교  학과  이름으로 구분하기
정보 = s.split()
print(정보, '학생이름:', 정보[2])
````

````python
#1. 문자열에서 분리해서 항목에 넣어보기
s = '인천대학교,  국어국문학과 ,  홍길동  , 90 ' 
정보 = s.split(',')
대학 = 정보[0]
학과 = 정보[1]
이름 = 정보[2]
성적 = 정보[3]
print(대학, 학과, 이름, 성적)
# 문제점 1 : 쓸데없는 빈칸 생성
학과 = 학과.strip()
이름 = 이름.strip()
성적 = 성적.strip()
print(len(대학), len(학과), len(이름), len(성적))
# 문제점 2 : 자료형
성적 = int(성적)
print(type(성적))
print(대학, 학과, 이름, 성적)
````

## find()
- 문자 x를 입력하면 그 문자가 존재하는 인덱스를 반환

- 찾는 문자가 없으면 **-1**을 반환
- 찾고싶은 문자가 여러개인경우 **맨 처음 **찾는 인덱스 반환
- 문자열.find(x, start, end)
```
s = '문제 해결을 위한  파이썬' 
s.find('문')
```
- 결과
    - ['문제', '해결을', '위한', '파이썬']

````python
s = '문제 해결을 위한  파이썬 해' 
s.find('해', 5, 10)
````

````python
s = 'HaaaaHaaaaaHa  ' 
s.find('H', 4, 15)
````

## 혼자해보기


date = '2021-12-25 토요일'
- 위 date에서 월(month)만 골라내 숫자형태로 변수 mon에 저장하시오.

````python
date = '2021-12-25 토요일'
# 공백을 기준으로 split
d1 = date.split()
# -을 기준으로 split
d2 = d1[0].split('-')
# split의 결과는 리스트이므로 인덱스 사용 가능
mon = d2[1]
# split의 결과는 문자이므로 숫자로 변환해야 함
mon = int(mon)
mon
````

````python
date = '2021-01-25'
# 월만 골라내기 - 코드를 간결하게 작성
mon = int(date.split('-')[1])
mon
````

위 date에서 일(day)만 숫자 형태로 골라내 변수에 저장하시오.
