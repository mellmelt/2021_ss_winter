# 반복문
컴퓨터 프로그래밍에서 반복문(영어: iteration)은 제어문중 하나로, 프로그램 소스 코드내에서 특정한 부분의 코드가 반복적으로 수행될 수 있도록 하는 구문이다.[위키백과]

- 컴퓨터는 똑같은 일을 반복 처리작업을 잘 한다.
- 반복구조를 사용하면 효율적으로 처리가 가능

    - while 문 - 어떤 조건이 **만족(True)** 되는 동안, 반복을 계속하는 구조
    - for 문 - ***정해진 횟수***만큼 반복하는 구조

### 문제 1
'파이썬'을 5번 출력

````python
# 반복문없는 출력
print('python')
print('python')
print('python')
print('python')
print('python')
````

````python
# 반복문없는 출력
print(1)
print(2)
print(3)
print(4)
print(5)
````

````python
# 반복문
count = 0

while count < 5: # count<6 보다는 기준이 되는 '5'로 표현하는 것이 좋다
  count += 1 # count 값을 먼저 더하고 나서 print 하기 때문에 5까지만 출력하고 싶다면 위에서 count < 5 이런식으로 초기값/조건문의 수정이 필요할 때도 있다
  print(count)
  # count += 1
````

### 문제 5

### 문제 7

````python
i = 1
hap = 0 # 합 초기화

while i <= 10:
  hap = hap + i # hap 초기값 = 0. i = 1. ==> hap = 1. # i=3일 때 hap = 1+2+3=6
  i += 1
print(hap)
````

### 문제 8

### 문제 10
이름을 입력하고, 문자 ‘q’를 입력하면 프로그램을 종료

````python
name = '' # 문자 초기화
while name != 'q':
  name = input('당신의 이름을 입력하세요 : ')
  print('당신의 이름은', name, '입니다.')
print('프로그램 종료')
````

## 무한 루프와 break
- 반복부분이 언제 끝나는지 정하지 못하는 경우 
- 조건 부분을 무조건(True)로 정하는 것을 무한 루프라 한다.
- 반복영역을 끝내기 위한 키워드 break

### 문제 12 
**반복문에서 탈출**
<br>특별히 정한 값으로 반복문 탈출
<br> 이름을 입력하고, 문자 ‘q’를 입력하면 프로그램을 종료

````python
while True:
  name = input('당신의 이름을 입력하세요 : (종료 : q) ')
  if name == 'q':
    break
  print('당신의 이름은', name, '입니다.')
print('프로그램 종료')
````

### 문제 13 
**무한 반복문에서 탈출**
<br>특별히 정한 값으로 반복문 탈출
<br> 1에서 5까지 출력


````python
i = 1

while True:
  print(i)
  if i == 5: break # break문을 바로 옆에 써도 괜찮다
  i += 1
````

### 문제 14 
**무한 반복문에서 탈출**
<br>■ 구구단 프로그램
<br>숫자를 입력하고 입력한 단을 출력하는 구구단 프로그램 
<br> 탈출 조건은 **숫자 0**이 입력되면 종료

````python
while True:
  dan = int(input('구하려는 단을 입력 : (종료: 0)'))
  if dan == 0 : break
  j = 2
  while j <= 9:
    print(dan, '*', j, '=', dan * j)
    j += 1
print('프로그램 종료')
````

### 문제 15  [혼자해보기]
 **무한 반복문에서 탈출**
<br>특별히 정한 값으로 반복문 탈출
<br> 이름을 입력후 입력 문자를 출력한다.
- [출력]


```
이름을 입력: 홍길동
홍길동님 안녕하세요. 1번째 손님입니다.

이름을 입력: 박문수 
박문수님 안녕하세요. 2번째 손님입니다.

이름을 입력: q
손님은 2명입니다.

[종료] 이름입력시‘q' 또는 'Q'를 입력하면 종료
```





````python
a = 0

while True:
  name = input('이름을 입력(종료 : q 또는 Q) :')
  if name == 'q' or name == 'Q': break
  a += 1
  print(name, '님 안녕하세요.', str(a) + '번째 손님입니다.')
print('손님은', str(a) + '명입니다.')
````
