## 형식
- **<함수이름>** : 명칭으로 변수의 이름 생성규칙과 같다.
- **<매개변수ㅡ리스트>** : 콤마로 구분된 변수 목록으로 매개변수 = value로 설정하여 매개변수의 기본 값을 가진다. 기본값은 함수 호출 할 때 인수를 생략하는 경우 매개변수의 값이 된다. 
-**<수식_리스트>** : 함수가 반환할 값이다. 두 개이상의 값을 튜플 자료형식으로 반환한다.
- **return문**: 생략이 될 수 있고, 생략하면 None값이 반환된다.

```python
def hello():
  # 여러 문장 한꺼번에 들여쓰기 하기 : ctrl + ]
  print('Hi!!')
  print('pyhon')
  return # 함수 종료
  print('Programming')
hello()
```

```python
hello()
print('프로그램 시작')
hello()
print('함수실습 준비중')
hello()
print('함수수업 끝 ')
```
