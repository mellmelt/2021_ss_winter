## **재고금액** 구하기

- 제품의 재고수량과 금액을 이용하여 제품별 재고금액을 계산하여 출력해본다.

[출력형식]


```
태양 오렌지 주스	8000	39	312000
태양 레몬 시럽	1900	17	32300
태양 체리 소스	1000	13	13000
신한 복숭아 시럽	2200	53	116600
신한 파인애플 시럽	2100	12	25200
대양 특선 블루베리 잼	2500	120	300000
```




### 프로그램(1) - 기본작성

- 재고금액구하기
    -  제품단가 * 재고수량

````python
for 행리스트 in data: # ['태양 오렌지 주스', '8000', '39', 'N', '4']
  제품명 = 행리스트[0]
  제품단가 = int(행리스트[1])
  재고수량 = int(행리스트[2])
  
  재고금액 = 제품단가 * 재고수량

  print(제품명, 제품단가, 재고수량, 재고금액, sep = '\t')
````

### 프로그램(2) - 함수를 정의후 작성

#### 재고금액을 계산하는 함수 
- 함수명 : fn_재고금액
- 전달값: 단가, 수량
- 반환값 : 단가 * 수량 



````python
#함수 정의
def fn_재고금액(단가, 수량):
  결과 = 단가 * 수량
  return 결과
````

````python
#함수 호출
# 반환하는 값이 있다면 꼭 print문으로 호출하자
print(fn_재고금액(5000, 6))

금액 = fn_재고금액(10000, 4)
print(금액)
````

## 할인률 구하기

### 할인률 정하는 규칙

- 재고수량이 
    - 100개 이상인경우 할인률 40%
    - 50개 이상인경우 할인률 30%
    - 30개 이상인경우 할인률 20%
    - 10개 이상인경우 할인률 10%
    - 10개미만인경우 할인률 5%

- 함수명: fn_할인률
- 전달값 : 재고수량
- 반환값 : 할인률





````python
#함수 정의
def fn_할인률(재고수량):
  if 재고수량 >= 100:
    할인률 = '40%' # 실제 할인률 계산하려면 할인율 = 0.4 이런식으로 작성해야 함
  elif 재고수량 >= 50:
    할인률 = '30%'
  elif 재고수량 >= 30:
    할인률 = '20%'
  elif 재고수량 >= 10:
    할인률 = '10%'
  else:
    할인률 = '5%'
  return 할인률
````

````python
#함수 호출
print(fn_할인률(35))
````

````python
#data에서 할인률 구하기
for 행리스트 in data:
  재고수량 = int(행리스트[2])
  할인률 = fn_할인률(재고수량)
  print(행리스트, 할인률)
````
