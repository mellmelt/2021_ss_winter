## range( ) 함수
- 지정한 범위안에서 정수를 생성할 수 있는 순회자(iterator)를 반환.
- 전달인자는 반드시 정수이어야 한다.
- 형식 :  range(시작값 , 종료값+1 , 증가값)

````python
print(list(range(0, 5, 1))) # range(5), 시작값/증가값 생략 가능. 생략 시 각각 0, 1로 지정됨
print(list(range(5))) # list 함수에 넣어줘야 결과 출력 가능
print(list(range(2, 7)))
print(list(range(2, 10, 2)))
print(list(range(2, 11, 2)))
````

## [문제 1]

'파이썬'을 5번 출력

````python
#while문으로
i = 0
while i < 5:
  print('파이썬')
  i += 1
print('프로그램 종료')
````

````python
#for문으로
for i in range(5): # [0, 1, 2, 3, 4] # 변수 i가 range(5)에 있는 값을 하나씩 받음
  print('파이썬')
print('프로그램 종료')
````

````python
list(range(5))
````

## [문제 2]

1에서 5까지 숫자를 출력
    - [출력]
```
    1
    2
    3
    4
    5


````python
#while문으로 
i = 0
while i <= 5:
  print(i)
  i += 1
print('프로그램 종료')
````

````python
#for문으로 
for i in range(1, 6): # [1, 2, 3, 4, 5]
  print(i)
print('프로그램 종료')
````

## [문제 3]

1에서 5까지 숫자를 출력
- [출력]
    <br>1 2 3 4 5

````python
#while문으로 
i = 1
while i <= 5:
  print(i, end = ' ') # end 속성
  i += 1
print('프로그램 종료')
````

````python
#for문으로 
for x in range(1, 6):
  print(x, end = ' ') # end 속성
````

## [문제 4]

출력형식으로 작성
- [출력]
```
     1 - 파이썬
     2 - 파이썬
     3 - 파이썬
     4 - 파이썬
     5 - 파이썬
    

````python
#while문으로  
i = 1
while i <= 5:
  print(i, '- 파이썬') # end 속성
  i += 1
print('프로그램 종료')
````

````python
#for문으로 

for 수 in range(1, 6):
  print(수, '- 파이썬')
````

## [문제 5]

구구단 5단 출력하기
- [출력]
<br>5 * 2 = 10
<br>5 * 3 = 15
<br> ...
<br>5 * 9 = 45

````python
#구구단 작성하기
list(range(2, 10))
# 2...9
````

````python
# 구구단 작성하기
for j in range(2, 10):
  print('5 *', j, '=', 5 * j)
````

## [문제 6] - 혼자해보기



구구단의 값을 입력받은후 결과를 출력

- [출력]
<br> 구하려는 단을 입력: **5**
<br>5 * 2 = 10
<br>5 * 3 = 15
<br> ...
<br>5 * 9 = 45

````python
# 구구단 작성하기
i = int(input('구하려는 단을 입력:'))
for j in range(2, 10):
  print(i, '*', j, '=', i * j)
````

## [문제 7]

 1에서 10까지 숫자의 합 출력하기
- [출력]
```
합 = 55


````python
#1에서 10까지 숫자의 합 출력
list(range(1, 11))
````

````python
hap = 0
for 값 in range(1, 11):
  hap = hap + 값 # hap += 값
print(hap)
````
