<br>사용자에게 구입 금액을 물어보고 회원등급을 출력하는 프로그램을 작성해보자.


```

```

### 문제4

[문제] 인터넷쇼핑몰 등급 프로그램

<br>      100,000원 미만이면 ’기본회원‘
```
출력 결과1)

````python
구입금액 = int(input('구입 금액을 입력하시오: '))

if 구입금액 >= 100000:
  print('회원등급이 우수회원입니다.')
else:
  print('회원등급이 기본회원입니다.')
print('구매 감사합니다.')
````

```
출력 결과1)--------------------
구입 금액을 입력하시오: 27000
쿠폰(Cash3000 또는 Cash5000) 선택 : Cash3000

결제금액 : 24000

## if ~ elif ~ else 문


Python-if-elif-ladder.jpg

````python
# 여러개의 조건을 적용하고 싶을 때 elif
score = 50
grade = ''

if score >= 90:
  grade = 'A'
elif score >= 80: # 89 >= score >= 80
  grade = 'B'
elif score >= 70:
  grade = 'C'
elif score >= 60:
  grade = 'D'
else:
  grade = 'F'

print(score, grade)
````

### 혼자해보기(3)

회원등급 정하기

회원의 포인트점수가

- 800점 초과이면 '최우수'
- 501~800점 이면 '우수'
- 301~500점 이면 '보통'
- 0  ~ 300점 이면 '일반'



````python
포인트 = int(input('회원의 포인트점수 : '))
회원등급 = ''

if 포인트 > 800:
  회원등급 = '최우수'
elif 포인트 >= 501:
  회원등급 = '우수'
elif 포인트 >= 301:
  회원등급 = '보통'
else:
  회원등급 = '일반'

print('회원의 포인트점수:', 포인트, '회원등급:', 회원등급)
````

### 문제6

이름과 성적을 입력받은후<br>
[평가기준]
- 90점이상 "최우수 합격"
- 60점이상 "합격"
- 60점미만 "불합격"<br>

<br>[출력형식]<br>
홍길동님은 80점으로 합격입니다.

````python
name = input('이름을 입력 : ')
score = int(input('성적 입력 : '))
result = ''

if score >= 90:
  result = '최우수 합격'
else:
  if score >= 60:
    result = '합격'
  else:
    result = '불합격'

print('수고하셨습니다.') # 들여쓰기를 잘못 하면 90점 이상일 때 print 실행이 안됨 (else에 포함되어서)
print(name, score, result)
````

```
