## 함수정의(3) - 반환값이 있는 경우

- return문: 생략이 될 수 있고, 생략하면 None값이 반환된다.

- 함수호출시 처리방법
1. 변수에 반환값을 **저장**한다
2. 반환값을 print() 함수에 담아서 반환값을 **확인**한다.


````python
# 함수정의
def hello2():
  print('Hi!!')
  print('Python')
  return
  print('Programming')
  print('문제를 위한 파이썬')
````

````python
# 함수호출
# 반환값이 없는 경우.
print(hello2())
````

````python
# 함수정의  - sum1()
# 0에서 9까지 합을 출력하는 함수
def sum1():
  hap = 0
  for i in range(10):
    hap += i
  print(hap)

 # 함수호출   
sum1()
````

````python
# 함수정의 - sum2()
# 0에서 9까지 합을 반환하는 함수
def sum2():
  hap = 0
  for i in range(10):
    hap += i
  return hap
    
# 함수호출 1 
# sum2() # 반환값이 있다면 이렇게 호출하면 안됨
res = sum2() # 결과값을 재사용하려면 변수에 담기
res = res + 10
print(res)
# 함수호출 2 
print(sum2()) # print 함수로 값 반환
````

````python
# 함수정의 sum3()
# 0에서 x까지 합을 반환하는 함수
    # 함수 호출할때 전달값이 있는경우
def sum3(x):
  hap = 0
  for i in range(x+1):
    hap += i
  return hap
    
# 함수호출 1 
print(sum3(10))
# 함수호출 2 
res = sum3(20)
print(res)
````

#[혼자해보기 - 2]

함수이름 : 띠구하기<br>
전달인자 : 태어난년도 <br>
반환값 : 12간지중 동물 <br>

예) 1995년생은 돼지띠입니다.
반환하는 값 : 돼지

1995을 12로 나누면 나머지가 3 인경우 돼지

```
[참고]12간지 동물
12로 나누었을때 0 이면 원숭이
12로 나누었을때 1 이면 닭
12로 나누었을때 2 이면 개
12로 나누었을때 3 이면 돼지
12로 나누었을때 4 이면 쥐
12로 나누었을때 5 이면 소
12로 나누었을때 6 이면 호랑이
12로 나누었을때 7 이면 토끼
12로 나누었을때 8 이면 용
12로 나누었을때 9 이면 뱀
12로 나누었을때 10 이면 말
12로 나누었을때 11 이면 양

```




````python
def 띠구하기(태어난년도):
  if 태어난년도 % 12 == 0:
    띠 = '원숭이'
  elif 태어난년도 % 12 == 1:
    띠 = '닭'
  elif 태어난년도 % 12 == 2:
    띠 = '개'
  elif 태어난년도 % 12 == 3:
    띠 = '돼지'
  elif 태어난년도 % 12 == 4:
    띠 = '쥐'
  elif 태어난년도 % 12 == 5:
    띠 = '소'
  elif 태어난년도 % 12 == 6:
    띠 = '호랑이'
  elif 태어난년도 % 12 == 7:
    띠 = '토끼'
  elif 태어난년도 % 12 == 8:
    띠 = '용'
  elif 태어난년도 % 12 == 9:
    띠 = '뱀'
  elif 태어난년도 % 12 == 10:
    띠 = '말'
  elif 태어난년도 % 12 == 11:
    띠 = '양'
  else:
    띠 = ' '
  return  띠

birth = int(input('태어난 년도를 입력하세요 : '))
띠동물 = 띠구하기(birth)
print('당신의 태어난 해의 동물은', 띠동물,'입니다')
````
