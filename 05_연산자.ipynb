##문제1


 반지름을 입력받은후
<br> 원의 면적을 계산하는 수식은 반지름의 제곱 후 3.14을 곱한다<br>
<br> [출력형식]
<br> 반지름이 *5*인 원의 면적은 *78.5*입니다.

```python
# 수식 : 반지름(r) ** 2 * 3.14
# (1) : 값을 입력받거나(반지름) 변수에 선언(면적)
r = int(input('반지름을 입력하세요 : '))
# (2) : 원하는 값을 구하기위해 필요한 수식, 알고리즘
면적 = r ** 2 * 3.14
# (3) : 처리한 결과를 화면에 출력(출력 형식은 사용자 입장에서)
print('반지름이', str(r) + '인 원의 면적은', str(면적) + '입니다.')
```

##문제2


두개의 정수 a,b 을 입력받은후
<br> a를 b로 나눈 몫과 나머지를 출력한다.<br>
<br> [출력형식]
<br> a/b의 몫 : 4
<br> a/b의 나머지 : 2

##문제3

피트를 미터로 변환하기 **변수 사용**
<br> feet (피트)로 입력받은 값을 meter(미터)로 변환해주는 프로그램
<br> 참고) 1피트는 0.305미터이다. <br>
<br> [출력형식]
<br> 피트의 값을 입력하세요 : 16.5 
<br> 16.5 피트는 5.032 미터 입니다.

```python
f = float(input('피트의 값을 입력하세요: '))
m = round(f / 3.2787, 3) # round 함수로 원하는 만큼 소수점 반올림을 할 수 있다

print(f, '피트는', m, '미터 입니다.')
```
